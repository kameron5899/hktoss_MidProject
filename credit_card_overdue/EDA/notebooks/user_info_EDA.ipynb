```python
import pandas as pd
df1 = pd.read_csv('../../data/modified/EDA_completed.csv')
df2 = pd.read_csv('../../data/modified/user_info.csv')
```

```python
df1['발급회원번호'] = df1.index
df1.info()
```

# df1과 df2 병합

```python
# '발급회원번호' 열을 기준으로 데이터프레임 정렬
df1_sorted = df1.sort_values('발급회원번호').reset_index(drop=True)
df2_sorted = df2.sort_values('발급회원번호').reset_index(drop=True)

# '발급회원번호'를 기준으로 concat 사용하여 열 단위로 병합
df = pd.concat([df1_sorted, df2_sorted], axis=1)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df = df.drop(df[df['최초한도금액'] == 0].index)
```

```python
df.info()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

## 연령 컬럼 숫자형 데이터로 변환


```python
# df['연령'] = df['연령'].str.replace(r'\D','', regex=True).astype(int)
df.head()
```

```python
df['최상위카드등급코드'] = df['최상위카드등급코드'].str.replace('_', '0').astype(int)
df.head()
```

# 의미 파악이 어려운 컬럼 삭제

```python
df = df.drop(columns=['최상위카드등급코드'])
```

```python
# 삭제할 컬럼 리스트
columns_to_drop = ['RV신청일자','RV약정청구율','RV최소결제비율','수신거부여부_TM',
                   '수신거부여부_DM', '수신거부여부_메일', '수신거부여부_SMS',
                   '가입통신회사코드', '마케팅동의여부', '_2순위카드이용금액',
                   '_2순위카드이용건수', '_2순위신용체크구분']

# 여러 컬럼 삭제
df = df.drop(columns=columns_to_drop)
```

```python
df = df.dropna(subset=['_1순위신용체크구분'])
df['_1순위신용체크구분'].value_counts(dropna = False)
```

```python
# '_1순위카드이용건수' 컬럼에서 0 미만인 행을 추출
negative_usage_df = df[df['_1순위카드이용건수'] < 0]

# 결과 출력
negative_usage_df.head()
```

```python
df = df.loc[:, ~df.columns.duplicated()]
df.head()
```

# 날짜형 datetime형으로 변경

```python
df = df.dropna(subset=['최종유효년월_신용_이용가능', '최종유효년월_신용_이용', '최종카드발급일자'])


```

```python
df['최종유효년월_신용_이용가능'] = df['최종유효년월_신용_이용가능'].astype(int)
df['최종유효년월_신용_이용'] = df['최종유효년월_신용_이용'].astype(int)
df['최종카드발급일자'] = df['최종카드발급일자'].astype(int)

df['기준년월'] = pd.to_datetime(df['기준년월'].astype(str), format='%Y%m')
df['입회일자_신용'] = pd.to_datetime(df['입회일자_신용'].astype(str), format='%Y%m%d')
df['최종유효년월_신용_이용가능'] = pd.to_datetime(df['최종유효년월_신용_이용가능'].astype(str), format='%Y%m')
df['최종유효년월_신용_이용'] = pd.to_datetime(df['최종유효년월_신용_이용'].astype(str), format='%Y%m')
df['최종카드발급일자'] = pd.to_datetime(df['최종카드발급일자'].astype(str), format='%Y%m%d')
```

# 범주형 데이터를 수치형 데이터로 변환


```python
categorical_columns = ['카드론동의여부', 'RV전환가능여부']

for col in categorical_columns:
    if df[col].dtype == 'object':  # 문자열 데이터에 대해서만 적용
        df[col] = df[col].apply(lambda x: 1 if x in ['Z', 'Y'] else 0)
```

```python
df['연령'] = df['연령'].str.replace(r'\D','', regex=True).astype(int)
```

```python
df = df.drop(columns=['VIP등급코드'])
```

```python
df = df.drop(columns=['Life_Stage'])
```

```python
df = df.drop(columns=['거주시도명', '직장시도명'])
```
